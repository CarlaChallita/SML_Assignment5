```r
library(ISLR)
data(Auto)
library(e1071)
```

```r
#Part a
Auto['binary'] <- ifelse(Auto$mpg > median(Auto$mpg), 1, 0)
```

```r
#Part b
set.seed(1)
model_pred <- tune(svm, binary ~ ., data = Auto, kernel = "linear", ranges = list(cost = c(0.01, 0.1, 1, 5, 10, 100, 1000)))
summary(model_pred)
#The best cost is the one with the lowest error, which is a cost of 1
```

```r
#Part c
set.seed(1)
model_pred <- tune(svm, binary ~ ., data = Auto, kernel = "polynomial", ranges = list(cost = c(0.01, 0.1, 1, 5, 10, 100), degree = c(2, 3, 4)))
summary(model_pred)
#For the polynomial kernel, the lowest error is obtained for a degree of 2 and a cost of 100.
```

```r
set.seed(1)
model_pred <- tune(svm, binary ~ ., data = Auto, kernel = "radial", ranges = list(cost = c(0.01, 0.1, 1, 5, 10, 100), gamma = c(0.01, 0.1, 1, 5, 10, 100)))
summary(model_pred)
#For the radial kernel, the lowest error is obtained for a gamma of 0.01 and a cost of 100.
```

```r
#Part a
set.seed(1)
sample_data <- sample(nrow(OJ), 800)
train <- OJ[sample_data, ]
test <- OJ[-sample_data, ]
```

```r
#Part b
support_vector <- svm(Purchase ~ ., data = train, kernel = "linear", cost = 0.01)
summary(support_vector)
#The support vector classifier creates 435 support vectors, out of which 219 belong to level CH and 216 belong to level MM.
```

```r
#Part c
train_pred <- predict(support_vector, train)
table(train$Purchase, train_pred)
```

```r
(75+65)/(420+65+75+240)
#The train error rate is 17.5%.
```

```r
test_pred <- predict(support_vector, test)
table(test$Purchase, test_pred)
```

```r
(33+15)/(153+15+33+69)
#The test error rate is 17.78%.
```

```r
#Part d
set.seed(2)
support_vector <- tune(svm, Purchase ~ ., data = train, kernel = "linear", ranges = list(cost = 10^seq(-2, 1, by = 0.25)))
summary(support_vector)
#The optimal cost is 1.778279
```

```r
#Part e
support_vector <- svm(Purchase ~ ., kernel = "linear", data = train, cost = support_vector$best.parameter$cost)
train_pred <- predict(support_vector, train)
table(train$Purchase, train_pred)
```

```r
(62+69)/(423+62+69+246)
#The train error rate is now 16.375%.
```

```r
test_pred <- predict(support_vector, test)
table(test$Purchase, test_pred)
```

```r
(29+12)/(156+12+29+73)
#The test error rate is now 15.18%.
```
